```python
df=sns.load_dataset('tips')
df
```

```python
#plotting graph between tip and bill to see the relationship
sns.relplot(x='total_bill',y='tip',data=df)
plt.show()
```

```python
#to plot a line graph
sns.relplot(x='total_bill',y='tip',data=df,kind='line')
plt.show()
```

```python
#if we want to check if meal is lunch or dinner 
#we can use diffrent dot for lunch and dinner
sns.relplot(x='total_bill',y='tip',data=df,hue='time')
plt.show()
```

```python
#rather than diffrent colors for lunch and dinner if we want to use diffrent symbol the "style"
sns.relplot(x='total_bill',y='tip',data=df,style='time')
plt.show()
```

```python
#now suppose if we want to check if person is smoker or not
sns.relplot(x='total_bill',y='tip',data=df,style='time',hue='smoker')
plt.show()
```

```python
#so the data isnt quite well visible 
#so if we want two diffrent graph for smokers and non smokers
sns.relplot(x='total_bill',y='tip',data=df,style='time',col='smoker')
plt.show()
```

```python
######################plotting with catagorical data
df2=sns.load_dataset('tips')
df2
```

```python
#by default strip plot
sns.catplot(x='day',y='total_bill',data=df2)
plt.show()
```

```python
#by default strip plot
sns.catplot(x='day',y='total_bill',data=df2)
plt.show()
```

```python
#by default strip plot
#by default jitter =1
#jitter control the deviation of the point
sns.catplot(x='day',y='total_bill',data=df2,kind='strip',jitter=0.25)
plt.show()
```

```python
#swarm plot
#smarm plot doesnt allow overlapping of markers...it adjust the point along catagorical axis
sns.catplot(x='day',y='total_bill',data=df2,kind='swarm',hue='smoker')
plt.show()
```

```python
#box plot
#shows 25th,median and 75th percentile
sns.catplot(x='sex',y='total_bill',data=df2,kind='box')
plt.show()
```

```python
#check if smoker or not
sns.catplot(x='sex',y='total_bill',data=df2,kind='box',hue='smoker')
plt.show()
```

```python
########33boxen plot
#shows 25th,median and 75th percentile+below 25th percentile+above 75th percentile
sns.catplot(x='sex',y='total_bill',data=df2,kind='boxen')
plt.show()
```
