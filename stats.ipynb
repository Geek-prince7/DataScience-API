```python
#startup from delhi and banglore
city=df[(df['CityLocation']=='Bangalore') | (df['CityLocation']=='bangalore') | (df['CityLocation']=='New Delhi')]
city
```

```python
population=city['AmountInUSD']
population.dropna(inplace=True)
population
```

```python
population=population.str.replace(',','')
population=pd.to_numeric(population)
samples=population.sample(100,random_state=1)
samples
```

```python
pop_max=population.max()
pop_min=population.min()
pop_avg=population.mean()


sample_avg=samples.mean()
sample_min=samples.min()
sample_max=samples.max()
```

```python
print(pop_max,"  ",sample_max," ",int(pop_max)-int(sample_max))
print(pop_min,"  ",sample_min," ",int(pop_min)-int(sample_min))
print(pop_avg,"  ",sample_avg," ",pop_avg-sample_avg)
```
