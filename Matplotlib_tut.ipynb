# Pie chart

```python
import matplotlib.pyplot as plt
#basic pie chart
arr=[2,4,6,5]
plt.pie(arr)
plt.show()
```

```python
#to make both axis equal or to make it circular
arr=[2,4,6,5]
plt.pie(arr)
plt.axis("equal")
plt.show()
```

```python
#to change color
arr=[2,4,6,5]
color=["blue","green","purple","red"]
plt.pie(arr,colors=color)
plt.axis("equal")
plt.show()
```

```python
# to add label
arr=[2,4,6,5]
label=["A","B","C","D"]
plt.pie(arr,labels=label)
plt.axis("equal")
plt.show()
```

```python
#to add title
arr=[2,4,6,5]
label=["A","B","C","D"]
plt.pie(arr,labels=label)
plt.axis("equal")
plt.title("Pie chart demo")
plt.show()
```

```python
#to print % 
arr=[2,4,6,5]
label=["A","B","C","D"]
#plt.pie(arr,labels=label,autopct="%.1f") #autopct is used to print % .1f - upto 1 decimal place
#to print % sigh also
plt.pie(arr,labels=label,autopct="%.2f%%")
plt.axis("equal")
plt.show()
```

```python
#to explode a peice or more
arr=[2,4,6,5]
label=["A","B","C","D"]
explode=[0.1,0,0.2,0] #min 0 max 1
plt.pie(arr,labels=label,explode=explode) #use explode 
plt.axis("equal")
plt.show()
```

```python
#by def graph is anticlockwise to change it
arr=[2,4,6,5]
label=["A","B","C","D"]
plt.pie(arr,labels=label,counterclock=False)
plt.axis("equal")
plt.show()
```

```python
#by def start angle is 0 degree to change it
arr=[2,4,6,5]
label=["A","B","C","D"]
plt.pie(arr,labels=label,startangle=100) #now A starts from 100 degree
plt.axis("equal")
plt.show()
```

```python
#basic histogram
hist_arr=[1,2,3,4,1,2,5,3,2,1,1,1,4]
plt.hist(hist_arr)
plt.show()
```

```python
b=[1,2,3,7,10,4,5,6,3,1,11,14,15,1,21,20,19]
plt.hist(b,edgecolor="black")
plt.show()
#here what happened it has clubbed some no and show their frequency 
#to range=(maxno-minno)/no of bins
#max=21 min=1 no of bins=10 range=20/10=2
```

```python
#to view betterb=[1,2,3,7,10,4,5,6,3,1,11,14,15,1,21,20,19]
import numpy as np
xt=np.arange(22)
plt.hist(b,edgecolor="black")
plt.xticks(xt)
plt.show()
#here what happened it has clubbed some no and show their frequency 
#to range=(maxno-minno)/no of bins
#max=21 min=1 no of bins=10 range=20/10=2
```

```python
#to set no of bins
import numpy as np
xt=np.arange(22)
plt.hist(b,bins=20,edgecolor="black")
plt.xticks(xt)
plt.show()
```

```python
#we can also set range of bins
import numpy as np
xt=np.arange(22)
plt.hist(b,bins=[1,3,6,8,11,15,20],edgecolor="black")
plt.xticks(xt)
plt.show()
```

```python
#to change x axis and y axis range
plt.hist(b,edgecolor="black")
plt.axis([0,25,0,6])
plt.show()
```

3D graph

```python
####################################################################################
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()
ax = plt.axes(projection='3d')
ax = plt.axes(projection='3d')

# Data for a three-dimensional line
zline = np.linspace(0, 15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline, 'gray')

# Data for three-dimensional scattered points
zdata = 15 * np.random.random(100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Greens');
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

```python
#####################################################################################
# Open and read data file as specified in the question
# Print the required output in given format
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv("amazon_jobs_dataset.csv", encoding = 'utf-8')
df = df['location']
city = df.str.split(", ",expand = True)
India = city[city[0] == "IN"]
city1= India[2]
freq = city1.value_counts(ascending =True)
x = freq.index
y = freq.values
plt.pie(y,labels = x,autopct = "%.2f%%")
plt.axis('equal')
plt.show()
a = np.true_divide(y,y.sum(axis=0, keepdims=True))*100
for i in range(len(x)):
    print(x[i],format(a[i],'.2f'))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("amazon_jobs_dataset.csv")
arr=list(df["location"])
d={}
for i in arr:
    val=i.split(', ')
    if val[0]=='IN':
        d[val[2]]=d.get(val[2],0)+1
count_arr=[]
city=[]
for i in d:
    city.append(i)
    count_arr.append(d[i])
city,count_arr

plt.pie(count_arr,labels=city,autopct="%.2f%%")
plt.show()
sumc=sum(count_arr)
for i in range(len(city)):
    print(city[i],format(count_arr[i]*100/sumc,".2f"))
        
```
