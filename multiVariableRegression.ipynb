```python
from sklearn import datasets 
```

```python
data= datasets.load_boston()
```

```python
X1=data.data
Y1=data.target
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(X1)
df.columns=data.feature_names
df
```

```python
header=data.feature_names
```

```python
for i in range(11):
    df[header[i]+' '+header[i+1]]=df[header[i]]*df[header[i+1]]
```

```python
df.head()
```

```python
X2=df.values
Y2=Y1
X2.shape
```

```python
from sklearn import model_selection as ms
```

```python
X1_tr,X1_te,Y1_tr,Y1_te=ms.train_test_split(X1,Y1,random_state=0)
X2_tr,X2_te,Y2_tr,Y2_te=ms.train_test_split(X2,Y2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
algo1=LinearRegression()
algo2=LinearRegression()
```

```python
algo1.fit(X1_tr,Y1_tr)
algo2.fit(X2_tr,Y2_tr)
```

```python
Y1_pred=algo1.predict(X1_te)
Y2_pred=algo2.predict(X2_te)
```

```python
score1_test=algo1.score(X1_te,Y1_te)
score1_train=algo1.score(X1_tr,Y1_tr)
score2_test=algo2.score(X2_te,Y2_te)
score2_train=algo2.score(X2_tr,Y2_tr)
```

```python
print('Training data score algo1:',score1_train,' test data score algo1 : ',score1_test)
print('Training data score algo2:',score2_train,' test data score algo2 : ',score2_test)
```
