```python
def score(y_pred,y_true):
    u=((y_true-y_pred)**2).sum()
    v=((y_true-y_true.mean())**2).sum()
    return 1-u/v
```

```python
def cost(X,Y,m):
    N=len(X)
    total_cost=0
    M=len(X[1,:])
    
    for i in range(N):
        x=X[i,:-1]
        y=Y[i]
        val=0
        for j in range(M-1):
            val+=m[j]*x[j]
        total_cost+=(1/N)*((y-val-1)**2)
    return total_cost
```

```python
def gd_repeat(X,Y,lr,m):
    M=len(X)
    N=len(X[1,:])
    m_=np.zeros(N)
    newm=np.zeros(N)
    for i in range(len(X)):
        x=X[i,:-1]
        y=Y[i]
        val=0
        for k in range(N-1):
            val+=m[k]*x[k]
        for j in range(N-1):
            m_[j]+=(-2/M)*(y-val-m[-1])*x[j]
        m_[-1]+=(-2/M)*(y-val-m[-1])
    for j in range(N-1):
        newm[j]=m[j]-lr*m_[j]
    return newm
```

```python
def gd(X,Y,learning_rate,no_iteration):
    m=np.zeros(len(X[1,:]))
    for i in range(no_iteration):
        m=gd_repeat(X,Y,lr,m)
        print(i,' ',cost(X,Y,m))
    return m
```
